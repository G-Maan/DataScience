```python
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns
sns.set(style="white", color_codes=True)
```

## Pokemon dataset

```python
pokemon = pd.read_csv('pokemon.csv')
pokemon_summary = pokemon.describe().transpose()
pokemon_summary
```

### Drop legendary and ID column

```python
pokemon = pokemon.drop(['Legendary','#'],1)
pokemon.head()
```

## Correlation and Covariance

```python
sns.heatmap(pokemon.corr(), annot=True)
```

```python
sns.heatmap(pokemon.cov(), annot=True)
```

## Plots of all attributes against each other

```python
p = sns.PairGrid(pokemon, hue='Generation')
p = p.map_diag(plt.hist)
p = p.map_offdiag(plt.scatter)
```

## Jointplots

```python
def draw_joinplot(col1, col2, data):
    g = sns.jointplot(x=col1, y=col2, data=data)
    
draw_joinplot(pokemon['Attack'], pokemon['Defense'], data=pokemon)
```

## Grouped by attributes

```python
g = sns.factorplot(x="Type 1", y="Attack", hue="Generation", size=20, data=pokemon, kind="bar")
g.despine(left=True)
```

```python
import random
import math
import scipy.stats as stats
import numpy as np

np.random.seed(123)
sample_size = 300
print('All pokemon mean:', pokemon.HP.mean())
sample_hp = np.random.choice(a=pokemon.HP, size=sample_size)
print('Sample pokemon mean:', sample_hp.mean())
print('Point estimation:', pokemon.HP.mean() - sample_hp.mean())
pokemon_std = pokemon.HP.std()
sample_mean = sample_hp.mean()
margin_error = pokemon_std / math.sqrt(sample_size)
confidence_interval = (sample_mean - margin_error, sample_mean + margin_error)
print('Confidence interval:', confidence_interval)
```
