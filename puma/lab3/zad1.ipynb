```python
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.animation as animation
from matplotlib.animation import FuncAnimation
from IPython.display import HTML
```

```python
def generate_samples(scale, samples, size):
    arr = {}
    for i in range(samples):
        arr[i] = np.random.exponential(scale=scale, size=size)
    return arr

def calculate_means(arrays):
    means = []
    for i in range(len(arrays)):
        mean = np.mean(arrays[i])
        means.append(mean)
    return means
```

```python
# zad 1 a
fig = plt.figure()
ax1 = fig.add_subplot(1,1,1)
ax1.set_ylim((0,1))

scale = 2
samples = 100
arr = {}
n = 0

def animate(i):
    global n
    n += 1
    arrays = generate_samples(scale, samples, n)
    means = calculate_means(arrays)
    ax1.cla()
    ax1.hist(means)

anim = animation.FuncAnimation(fig, animate, interval=100)
HTML(anim.to_jshtml())
```

```python
# zad 1 b
fig = plt.figure()
ax1 = fig.add_subplot(1,1,1)
ax1.set_ylim((0,1))

scale = 2
samples = 100
arr = {}
n = 0

def generate_binomial_samples(n, prob, samples):
    arr = {}
    for i in range(samples):
        arr[i] = np.random.binomial(n, prob, samples)
    return arr

def calculate_sum(arrays):
    count = []
    for i in range(len(arrays)):
        c = np.sum(arrays[i])
        count.append(c)
    return count

def animate(i):
    global n
    n += 1
    arrays = generate_binomial_samples(1, 0.5, n)
    count = calculate_sum(arrays)
    ax1.cla()
    ax1.set_ylabel('Count')
    ax1.set_xlabel('Trials')
    ax1.hist(count)

    
anim = animation.FuncAnimation(fig, animate, interval=100)
HTML(anim.to_jshtml())
```
