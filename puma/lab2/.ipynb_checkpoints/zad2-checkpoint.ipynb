```python
# zad1 intro
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.mlab as mlab

mu = 0
sigma = 0.02
samples = 100
interval = 4

def normal_distribution(x, mu, sigma):
  element_1 =  1/ np.sqrt(2 * np.pi * (sigma**2))
  element_2 = np.e ** -((x - mu) ** 2 / (2 * sigma ** 2))
  return element_1 * element_2
  
x = np.linspace(mu - interval * sigma, mu + interval * sigma, samples)
npdf = mlab.normpdf(x, mu, sigma)
plt.plot(x, npdf)

def norm_distribution_pdf(mu, sigma, interval):
  x = np.linspace(mu - interval * sigma, mu + interval * sigma, 1000)
  return x, mlab.normpdf(x, mu, sigma)

plt.show()
```

```python
#zad 1 a
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.mlab as mlab

mu = 4
sigma = 2
samples = 10
interval = 4

x_0, pdf_0 = norm_distribution_pdf(mu, sigma, interval)

rand0 = np.random.normal(mu, sigma, samples)

mu_R = np.sum(rand0) / rand0.size
var_R = np.array([(mu_R - el) ** 2 for el in rand0]).sum() / rand0.size
sd_R = np.sqrt(var_R)

print('mu={}\nvar={}\nsd={}'.format(mu_R, var_R, sd_R))

x, pdf_R = norm_distribution_pdf(mu_R, sd_R, interval)
plt.plot(x_0, pdf_0, 'b')
plt.plot(x, pdf_R, 'r')
plt.show()
```

```python
#zad1 b
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.mlab as mlab

samples = 100
interval = 4

mu_1 = 4
sigma_1 = 2

mu_2 = 3
sigma_2 = 5

x_1, pdf_1 = norm_distribution_pdf(mu_1, sigma_1, interval)
rand_1 = np.random.normal(mu_1, sigma_1, samples)

x_2, pdf_2 = norm_distribution_pdf(mu_2, sigma_2, interval)
rand_2 = np.random.normal(mu_2, sigma_2, samples)

plt.plot(x_1,pdf_1, 'b')
plt.plot(x_2,pdf_2, 'b')

rand_R = np.concatenate([rand_1,rand_2])

mu_R = np.sum(rand_R) / rand_R.size
var_R = np.array([(mu_R - el) ** 2 for el in rand_R]).sum() / rand_R.size
sd_R = np.sqrt(var_R)

print('mu={}\nvar={}\nsd={}'.format(mu_R, var_R, sd_R))

x_R, pdf_R = norm_distribution_pdf(mu_R, sd_R, interval)
plt.plot(x_R, pdf_R, 'r')
plt.show()
```

```python
#zad1 c
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.mlab as mlab

samples = 100
interval = 4

mu_1 = 4
sigma_1 = 2

mu_2 = 3
sigma_2 = 5

mu_3 = 1
sigma_3 = 0.5

mu_arr = [mu_1, mu_2, mu_3]
sigma_arr = [sigma_1, sigma_2, sigma_3]

def calculate_pdf_arrays(mu_arr, sigma_arr, interval):
    x_arr = []
    pdf_arr = []
    for mu,sigma in zip(mu_arr,sigma_arr):
        x,pdf = norm_distribution_pdf(mu,sigma,interval)
        x_arr.append(x)
        pdf_arr.append(pdf)
    return x_arr,pdf_arr

def get_random_arrays(mu_arr, sigma_arr, samples):
    rand_arr = []
    for mu,sigma in zip(mu_arr,sigma_arr):
        r = np.random.normal(mu,sigma,samples)
        rand_arr.append(r)
    return np.ndarray.flatten(np.array(rand_arr))

x_arr,pdf_arr = calculate_pdf_arrays(mu_arr,sigma_arr,interval)

rand_R = get_random_arrays(mu_arr,sigma_arr,samples)

for x,pdf in zip(x_arr,pdf_arr):
    plt.plot(x,pdf,'b')

mu_R = np.sum(rand_R) / rand_R.size
var_R = np.array([(mu_R - el) ** 2 for el in rand_R]).sum() / rand_R.size
sd_R = np.sqrt(var_R)

print('mu={}\nvar={}\nsd={}'.format(mu_R, var_R, sd_R))

x_R, pdf_R = norm_distribution_pdf(mu_R, sd_R, interval)
plt.plot(x_R, pdf_R, 'r')
plt.show()
```

```python
import scipy.special


def chance_using_probability(trials, success_prob):
  fail_prob = 1 - success_prob
  return (fail_prob) ** trials

print(chance_using_probability(9, 0.01))

def chance_using_binomial(trials, success_prob, number_of_successes):
  n_succ = number_of_successes
  n_fail = trials - number_of_successes

  return scipy.misc.comb(trials, n_succ) * success_prob ** (n_succ) * (1 - success_prob) ** (n_fail)

print(chance_using_binomial(9, 0.01, 0))
```
