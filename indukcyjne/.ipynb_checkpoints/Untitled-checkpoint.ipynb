```python
from sklearn import datasets
from sklearn.naive_bayes import GaussianNB
from sklearn.utils import shuffle
from sklearn.model_selection import KFold, StratifiedKFold, cross_val_score, train_test_split
import numpy as np
```

```python
iris = datasets.load_diabetes()

idx = np.random.permutation(len(iris.data))

X,Y = iris.data[idx],iris.target[idx]

s = np.arange(X.shape[0])
np.random.shuffle(s)

train_length = len(X)//10*8

x_train = X[:train_length,:]
y_train = Y[:train_length]
x_test = X[train_length:len(X),:]
y_test = Y[train_length:len(Y)]

print('X train shape:%s' % str(x_train.shape))
print('Y train shape:%s' % str(y_train.shape))
print('X test shape:%s' % str(x_test.shape))
print('Y test shape:%s' % str(y_test.shape))

for (x,y) in zip(X,Y):
    print(x)
    print(y)
```

```python
gnb = GaussianNB()
fit = gnb.fit(x_train, y_train)
y_pred = fit.predict(x_test)
errors = (x_test.shape[0],(y_test != y_pred).sum())
print("Number of mislabeled points out of a total %d points : %d"
      % errors)
print("Accuracy %% : %.2f" % (100-errors[1]/errors[0]*100))
```

```python
scores = cross_val_score(fit, iris.data, iris.target)
print("Accuracy: %0.2f (+/- %0.2f)" % (scores.mean(), scores.std() * 2))
```
